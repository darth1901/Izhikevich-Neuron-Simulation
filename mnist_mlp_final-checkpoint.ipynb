```python
import torch
import torchvision
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import tarfile
import os
```

```python
#Extraction, Loading and Preprocessing
    #A. File extraction
tar_path = 'mnist_png.tar.gz'
extract_path = './'
    
if not os.path.exists(os.path.join(extract_path, 'mnist_png')):
    with tarfile.open(tar_path, 'r:gz') as tar:
            tar.extractall(path=extract_path)
        
    #B. Preprocessing function (using the transform.Compose function of PyTorch which will perform a series of transformations on the data input)
transform = transforms.Compose([
transforms.Grayscale(num_output_channels=1), #declares every image as a greyscale image having only 1 color/channel
transforms.ToTensor(),                      #converts image from .png to tensor datatype
transforms.Normalize((0.1307,), (0.3081,)) #mean and sd for MNIST
])

    #C.Creating batches out of the dataset (of size 64)

#C.1. Defining a directory for the train and test set 
train_dir = os.path.join(extract_path, 'mnist_png/training')
test_dir = os.path.join(extract_path, 'mnist_png/testing')

#C.2.The ImageFolder function is a PyTorch function that will open a directory and then apply the required transformation. 
#Also as the train test directories have sub-directories containing the actual images. The ImageFolder function will automatically assign a label to each image of the directory as the name of that directory. So all images in '1' folder will be labelled as 1.
train_dataset  = ImageFolder(root=train_dir, transform=transform)
test_dataset = ImageFolder(root=test_dir, transform=transform)

#C.3. The DataLoader function is also a PyTorch function that will take as input 
b_size = 64

#batch sizes are taken as powers of 2 to fit the memory addresses better
train_loader = DataLoader(dataset=train_dataset, batch_size=b_size, shuffle=True)
test_loader = DataLoader(dataset=test_dataset, batch_size=b_size, shuffle=False)  
```

```python
#model
    #defining the nn (an mlp)    
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # Input layer (784) -> Hidden Layer 1 (256)
        self.fc1 = nn.Linear(784, 256) 
        # Hidden Layer 1 (256) -> Hidden Layer 2 (128)
        self.fc2 = nn.Linear(256, 128)
        # Hidden Layer 2 (128) -> Output Layer (10)
        self.fc3 = nn.Linear(128, 10)
        # Dropout Layer
        self.dropout = nn.Dropout(p=0.2)

    def forward(self, x):
        x = x.view(-1, 784)

        x = self.dropout(F.relu(self.fc1(x)))
        x = self.dropout(F.relu(self.fc2(x)))

        x = self.fc3(x)
        return x

```

```python
#Setup, Training, and Evaluation Loop 
model = Net()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
epochs = 15

# Lists to store metrics for plotting
train_losses = []
test_accuracies = []
epoch_nums = []

print("Starting training...")
for epoch in range(epochs):
    # --- Training Phase ---
    model.train()
    running_loss = 0.0
    for batch_idx, (images, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # --- Evaluation Phase (at the end of each epoch) ---
    model.eval()
    correct_test = 0
    total_test = 0
    with torch.no_grad():
        for images, labels in test_loader:
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total_test += labels.size(0)
            correct_test += (predicted == labels).sum().item()
    
    # --- Record Metrics ---
    epoch_loss = running_loss / len(train_loader)
    epoch_test_acc = 100 * correct_test / total_test
    
    train_losses.append(epoch_loss)
    test_accuracies.append(epoch_test_acc)
    epoch_nums.append(epoch + 1)
    
    print(f'Epoch {epoch+1}/{epochs} | Train Loss: {epoch_loss:.4f} | Test Accuracy: {epoch_test_acc:.2f}%')

print("\nTraining finished!")
```

```python
torch.save(model.state_dict(), 'mnist_mlp_final.pth')
print("MLP model saved to mnist_mlp_final.pth")
```

```python
# --- Plotting the Visualizations ---
plt.style.use('ggplot')
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Plot 1: Training Loss vs. Epoch
ax1.plot(epoch_nums, train_losses, marker='o', linestyle='-', color='r')
ax1.set_title('Training Loss vs. Epoch')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.grid(True)
ax1.set_xticks(epoch_nums)

# Plot 2: Test Accuracy vs. Epoch
ax2.plot(epoch_nums, test_accuracies, marker='o', linestyle='-', color='b')
ax2.set_title('Test Accuracy vs. Epoch')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.grid(True)
ax2.set_xticks(epoch_nums)
ax2.set_ylim(95, 100) # Zoom in on the high accuracy range

plt.tight_layout()
fig.savefig('training_visualizations.png')
plt.show()
```

```python
#testing script (for any model)
model = Net()
model.load_state_dict(torch.load('mnist_mlp_final.pth'))
model.eval()

correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        
        # Find the class with the highest score (the model's prediction)
        _, predicted = torch.max(outputs.data, 1)
        
        # Update the total and correct counts
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

# Calculate and print the final accuracy
accuracy = 100 * correct / total
print(f'\nAccuracy of the model on the {total} test images: {accuracy:.2f}%')
```
